```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.patches import ConnectionPatch
```

```python
x = np.arange(0, 10, 0.01)

y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.sin(x)

fig, ax = plt.subplots(3, 1, figsize=(5, 10), gridspec_kw=dict(hspace=0))
fig.tight_layout(pad=0)
ax[0].plot(x, y1)
ax[0].spines.top.set_visible(False)
ax[0].spines.bottom.set_visible(False)
ax[0].get_xaxis().set_visible(False)

ax[1].plot(x, y2)
ax[1].spines.top.set_visible(False)
ax[1].spines.bottom.set_visible(False)
ax[1].get_xaxis().set_visible(False)

ax[2].plot(x, y3)
ax[2].spines.top.set_visible(False)

cr1 = np.random.randint(0,1001)
cr2 = np.random.randint(0,1001)
cr3 = np.random.randint(0,1001)

pt1 = (x[cr1], y1[cr1])
pt2 = (x[cr2], y2[cr2])
pt3 = (x[cr3], y3[cr3])

con12 = ConnectionPatch(xyA=pt1, xyB=pt2, coordsA="data", coordsB="data", axesA=ax[0], axesB=ax[1], color="red")
con23 = ConnectionPatch(xyA=pt2, xyB=pt3, coordsA="data", coordsB="data", axesA=ax[1], axesB=ax[2], color="red")
ax[2].add_artist(con12)
ax[2].add_artist(con23)


plt.show()
```

```python
import matplotlib.pyplot as plt
from matplotlib.patches import ConnectionPatch
import numpy as np
plt.rcParams["figure.figsize"] = [7.00, 3.50]
plt.rcParams["figure.autolayout"] = True
fig = plt.figure()
ax1 = fig.add_subplot(121)
ax2 = fig.add_subplot(122)
x, y = np.random.rand(100), np.random.rand(100)
ax1.plot(x, y, 'd', c='r')
ax2.plot(x, y, 'd', c='r')
i = 10
xy = (x[i], y[i])
con1 = ConnectionPatch(xyA=xy, xyB=xy, coordsA="data", coordsB="data", axesA=ax2, axesB=ax1, color="blue")
j = 12
mn = (x[j], y[j])
con2 = ConnectionPatch(xyA=mn, xyB=mn, coordsA="data", coordsB="data", axesA=ax2, axesB=ax1, color="blue")
ax2.add_artist(con1)
ax2.add_artist(con2)
ax1.plot(x[i], y[i], 'bo', markersize=10)
ax2.plot(x[i], y[i], 'bo', markersize=10)
ax1.plot(x[j], y[j], 'bo', markersize=10)
ax2.plot(x[j], y[j], 'bo', markersize=10)
plt.show()
```
